2. Есть зависимость между продолжительностью работы в компании и семейным положением сотрудника.

##### У раздельно проживающих супругов (возможно это люди, находящиеся в длительной рабочей командировке, вахтовики) продолжительность работы выше, чем у не состоящих в браке, женатых, разведенных и вдовых. У вдовых сотрудников продолжительность работы самая низкая, несмотря на то, что их средний возраст не выше среднего возраста в остальных категориях, т.е. это явно НЕ предпенсионная категория.

```python
df_age_marital = pd.read_sql('SELECT ROUND(AVG(hr_dataset.age)) AS Средний_возраст \
FROM hr_dataset \
WHERE hr_dataset.maritaldesc = \'Separated\' \
UNION ALL \
SELECT ROUND(AVG(hr_dataset.age)) \
FROM hr_dataset \
WHERE hr_dataset.maritaldesc = \'Single\' \
UNION ALL \
SELECT ROUND(AVG(hr_dataset.age)) \
FROM hr_dataset \
WHERE hr_dataset.maritaldesc = \'Married\' \
UNION ALL \
SELECT ROUND(AVG(hr_dataset.age)) \
FROM hr_dataset \
WHERE hr_dataset.maritaldesc = \'Divorced\' \
UNION ALL \
SELECT ROUND(AVG(hr_dataset.age)) \
FROM hr_dataset \
WHERE hr_dataset.maritaldesc = \'Widowed\' \
', connect)

df_emp_marital = pd.read_sql('SELECT hr_dataset.maritaldesc AS "Семейное положение", \
                             ROUND(AVG(hr_dataset."Days Employed")) AS Стаж \
FROM hr_dataset \
GROUP BY "Семейное положение" \
ORDER BY Стаж DESC',connect)

df_emp_marital.merge(df_age_marital, how = 'left', left_index=True, right_index=True)
```

3. Есть зависимость между продолжительностью работы в компании и возрастом сотрудника.

##### Зависимости между продолжительностью работы в компании и возрастом сотрудника НЕ ВЫЯВЛЕНО.

```python
age_emp = pd.read_sql('SELECT ROUND(AVG(hr_dataset."Days Employed")) \
FROM hr_dataset WHERE hr_dataset.age <= 30 \
UNION ALL \
SELECT ROUND(AVG(hr_dataset."Days Employed")) \
FROM hr_dataset WHERE hr_dataset.age > 30 AND hr_dataset.age <= 40 \
UNION ALL \
SELECT ROUND(AVG(hr_dataset."Days Employed")) \
FROM hr_dataset WHERE hr_dataset.age > 40 AND hr_dataset.age <= 50 \
UNION ALL \
SELECT ROUND(AVG(hr_dataset."Days Employed")) \
FROM hr_dataset WHERE hr_dataset.age > 50 \
',connect)

print('Средний стаж до 30 лет:',age_emp.loc[0,'round'], ', от 30 до 40:',age_emp.loc[1,'round'], \
      ', от 40 до 50:',age_emp.loc[2,'round'], ', свыше 50:',age_emp.loc[3,'round'])
```
