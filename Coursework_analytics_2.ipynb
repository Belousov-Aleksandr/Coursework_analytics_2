```python
import numpy as np
import pandas as pd
from datetime import datetime, date, timedelta
from functools import partial
from statistics import mean, median

def strip(text):
    try:
        return text.strip()
    except AttributeError:
        return text
#импорт имеющихся csv в датасеты
df_hr_dataset = pd.read_csv('hr_dataset.csv',
                            index_col = 0, parse_dates = ['dob','Date of Hire','Date of Termination'],
                            dayfirst = False, encoding = "UTF-8", sep=',',
                           converters = {'Employee Name' : strip,
                                        'department' : strip})
df_production_staff = pd.read_csv('production_staff.csv',
                                  index_col = 0, parse_dates = ['Date of Hire','TermDate'],
                                  dayfirst = False, encoding = "UTF-8", sep=',',
                                 converters = {'Employee Name' : strip,
                                              'Department' : strip})
df_recruiting_costs = pd.read_csv('recruiting_costs.csv', sep=',', index_col = 0, encoding = "UTF-8")
df_salary_grid = pd.read_csv('salary_grid.csv', sep=',', index_col = 0, encoding = "UTF-8")

#дополнительная очистка данных
df_production_staff = df_production_staff.dropna(axis='index', how = 'any', subset=['Race Desc'])
df_hr_dataset.reset_index(drop=True, inplace=True)
df_production_staff.reset_index(drop=True, inplace=True)
df_recruiting_costs.reset_index(drop=True, inplace=True)
df_salary_grid.reset_index(drop=True, inplace=True)

df_hr_dataset.head()
```

```python
print('Работающих сотрудников -',df_hr_dataset[df_hr_dataset['Employment Status'] == 'Active']['Employee Name'].nunique(),
     ', уволенных по собственному желанию -',df_hr_dataset[df_hr_dataset['Employment Status'] == 'Voluntarily Terminated']['Employee Name'].nunique(),
     ', в отпуске -',df_hr_dataset[df_hr_dataset['Employment Status'] == 'Leave of Absence']['Employee Name'].nunique(),
     ', уволенных по негативу -',df_hr_dataset[df_hr_dataset['Employment Status'] == 'Terminated for Cause']['Employee Name'].nunique(),
     ', в процессе трудоустройства -',df_hr_dataset[df_hr_dataset['Employment Status'] == 'Future Start']['Employee Name'].nunique())

df_hr_dataset.groupby('Employment Status').agg({'empstatus_id': 'count'}).sort_values('empstatus_id', ascending = False)
```

```python
#очистка данных. удалим уволенных и еще не принятых
df_hr_dataset_active = df_hr_dataset[(df_hr_dataset['Employment Status'] == 'Active') | (df_hr_dataset['Employment Status'] == 'Leave of Absence')]
df_production_staff_active = df_production_staff[(df_production_staff['Employment Status'] == 'Active') | (df_production_staff['Employment Status'] == 'Leave of Absence')]

print('Компания состоит из',len(df_hr_dataset['department'].unique()),'подразделений:',
      ', '.join(df_hr_dataset['department'].unique()))

production_staff_1 = set(df_hr_dataset_active[df_hr_dataset_active['department'] == 'Production']['Employee Name'].tolist())
production_staff_2 = set(df_production_staff_active['Employee Name'].tolist())
multi_staff_count = len(production_staff_2 - production_staff_1)

df_staff_count = df_hr_dataset_active.groupby('department').agg({'Employee Name': 'count'}).sort_values('Employee Name', ascending = False)
df_staff_count.loc[df_staff_count.index.get_level_values(0)[0], 'Employee Name'] += multi_staff_count
df_staff_count.columns = ['Численность']
df_staff_count
```

```python
print('В производственном отделе (Production) работает:',len(production_staff_1 | production_staff_2),
      'сотрудников.\n')
print('Сотрудники:','; '.join(production_staff_2 - production_staff_1),
      'числятся в нескольких подразделениях.')
```

```python
stuff_active_count = df_hr_dataset_active['Employee Name'].nunique()
print('Гендерный состав сотрудников:',
      round((len(df_hr_dataset_active[df_hr_dataset_active['sex'] == 'Female']['sex']) / stuff_active_count) * 100),'% женщин,',
      round((len(df_hr_dataset_active[df_hr_dataset_active['sex'] == 'Male']['sex']) / stuff_active_count) * 100),'% мужчин.')
df_hr_gender = df_hr_dataset_active.groupby(['department','sex']).agg({'genderid':'count'})
df_hr_gender.columns = ['Количество']
df_hr_gender
```

```python
print('MAX возраст сотрудников:',max(df_hr_dataset_active['age'].tolist()),
      ', MIN возраст сотрудников:',min(df_hr_dataset_active['age'].tolist()),
      ', MEAN возраст сотрудников:',round(mean(df_hr_dataset_active['age'].tolist())),
     ', MEDIAN возраст сотрудников:',round(median(df_hr_dataset_active['age'].tolist())),'.')
df_hr_age = df_hr_dataset_active.groupby('department')['age'].agg(['min','mean','median','max'])
df_hr_age['mean'] = df_hr_gender['mean'].round(1)
df_hr_age
```

```python
#статистика по расам
df_hr_racedesc = df_hr_dataset_active.groupby(['department','racedesc']).agg({'Employee Name':'count'})
df_hr_racedesc.columns = ['Количество']
df_hr_racedesc
```

```python
#статистика по продолжительности работы в компании
df_hr_days_employed = df_hr_dataset.groupby('department')['Days Employed'].agg(['min','mean','median','max'])
df_hr_days_employed['mean'] = df_hr_days_employed['mean'].round(1)
df_hr_days_employed['median'] = df_hr_days_employed['median'].round(1)
df_hr_days_employed
```
